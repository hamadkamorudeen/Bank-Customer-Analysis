# Bank Customer Analysis

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# Load data
customer_df = pd.read_csv('Customer_Info.csv')
account_df = pd.read_csv('Account_Info.csv')
```

```python
# Drop duplicate 'Tenure' column in account_df
account_df = account_df.drop(columns=['Tenure'])
```

```python
# Merge datasets
df = pd.merge(customer_df, account_df, on='CustomerId', how='inner')
df = df.drop_duplicates(subset='CustomerId')
```

```python
# Clean euro symbols and convert to float
df['EstimatedSalary'] = df['EstimatedSalary'].replace('[€,]', '', regex=True).astype(float)
df['Balance'] = df['Balance'].replace('[€,]', '', regex=True).astype(float)
```

```python
# Age groupings
age_bins = [0, 25, 35, 45, 55, 65, 100]
age_labels = ['<25', '25-34', '35-44', '45-54', '55-64', '65+']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
# Drop CustomerId
df = df.drop(columns=['CustomerId'])
```

```python
# Clean up Geography names
df['Geography'] = df['Geography'].replace({
    'FRA': 'France', 'French': 'France'})
```

```python
# Drop NAs and duplicates
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
# Set plot style
sb.set(style="whitegrid")
plt.figure(figsize=(12, 6))
```

## Demographic Analysis

```python
# Distribution by geography, gender, and age group
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
sb.countplot(data=df, x='Geography', ax=axs[0])
sb.countplot(data=df, x='Gender', ax=axs[1])
sb.countplot(data=df, x='AgeGroup', ax=axs[2])
axs[0].set_title("Geography Distribution")
axs[1].set_title("Gender Distribution")
axs[2].set_title("Age Group Distribution")
plt.tight_layout()
plt.show()
```

```python
# Credit score by gender and geography
plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='Gender', y='CreditScore')
plt.title("Credit Score by Gender")
plt.show()

plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='Geography', y='CreditScore')
plt.title("Credit Score by Geography")
plt.show()
```

## Customer Engagement Patterns

```python
# Tenure vs Products, HasCrCard, IsActiveMember
plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='Tenure', y='NumOfProducts')
plt.title("Number of Products by Tenure")
plt.show()

sb.countplot(data=df, x='Tenure', hue='IsActiveMember')
plt.title("Active Membership across Tenure")
plt.show()
```

## Financial Performance

```python
# Credit Score by AgeGroup
plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='AgeGroup', y='CreditScore')
plt.title("Credit Score by Age Group")
plt.show()
```

```python
# Balance and EstimatedSalary comparison
plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='AgeGroup', y='Balance')
plt.title("Balance by Age Group")
plt.show()

plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='AgeGroup', y='EstimatedSalary')
plt.title("Estimated Salary by Age Group")
plt.show()
```

## Exited Customers Analysis

```python
# Compare exited vs non-exited customers
exit_df = df.copy()
exit_df['Exited'] = exit_df['Exited'].map({0: 'Active', 1: 'Exited'})

plt.figure(figsize=(10, 5))
sb.countplot(data=exit_df, x='AgeGroup', hue='Exited')
plt.title("Exited vs Active by Age Group")
plt.show()

plt.figure(figsize=(10, 5))
sb.boxplot(data=exit_df, x='Exited', y='CreditScore')
plt.title("Credit Score: Exited vs Active")
plt.show()

plt.figure(figsize=(10, 5))
sb.boxplot(data=exit_df, x='Exited', y='Balance')
plt.title("Balance: Exited vs Active")
plt.show()
```

## Customer Segmentation

```python
# High balance vs Low balance segmentation
threshold = df['Balance'].median()
df['BalanceSegment'] = df['Balance'].apply(lambda x: 'High' if x > threshold else 'Low')

plt.figure(figsize=(10, 5))
sb.countplot(data=df, x='BalanceSegment', hue='AgeGroup')
plt.title("Balance Segment by Age Group")
plt.show()

plt.figure(figsize=(10, 5))
sb.boxplot(data=df, x='BalanceSegment', y='EstimatedSalary')
plt.title("Salary by Balance Segment")
plt.show()
```

# Bank Customer Analysis

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline


# Load data
customer_df = pd.read_csv('Customer_Info.csv')
account_df = pd.read_csv('Account_Info.csv')

# Drop duplicate 'Tenure' column in account_df
account_df = account_df.drop(columns=['Tenure'])

# Merge dataframes on 'Customer_ID'
df = pd.merge(customer_df, account_df, on='CustomerId', how='inner')
df = df.drop_duplicates(subset='CustomerId')

# Clean euro symbols and convert to float
df['EstimatedSalary'] = df['EstimatedSalary'].replace('[€,]', '', regex=True).astype(float)
df['Balance'] = df['Balance'].replace('[€,]', '', regex=True).astype(float)

# Age groupings
age_bins = [0, 25, 35, 45, 55, 65, 100]
age_labels = ['<25', '25-34', '35-44', '45-54', '55-64', '65+']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Drop CustomerId
df = df.drop(columns=['CustomerId'])

# Clean up Geography names
df['Geography'] = df['Geography'].replace({
    'FRA': 'France', 'French': 'France'})

# Drop NAs and duplicates
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Set plot style
sb.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Demographic Analysis
# Distribution by geography, gender, and age group
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
sb.countplot(data=df, x='Geography', ax=axs[0])
sb.countplot(data=df, x='Gender', ax=axs[1])
sb.countplot(data=df, x='AgeGroup', ax=axs[2])
axs[0].set_title("Geography Distribution")
axs[1].set_title("Gender Distribution")
axs[2].set_title("Age Group Distribution")
plt.tight_layout()
plt.show()
```
